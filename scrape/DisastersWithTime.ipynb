```python
from parse_table import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import operator as o
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
from json import dumps
```

```python
t = Table.from_file("/home/mnewton/projects/d3data/app/data/emdat-advandced-search.csv", col_delim=",")
years_dict = t.get_classified_columns(lambda row: row['year'])
years_group = [(k,map(sum, zip(*[x[3:] for x in v]))) for k,v in t.get_classified_columns(lambda row: row['year']).iteritems()]
time = [year for year,vals in years_group]
```

```python
years_list = dict( (k, [dict(zip(t.labels, x)) for x in v]) for k,v in t.get_classified_columns(lambda row: row['year']).iteritems())
with open("/home/mnewton/projects/d3data/app/data/emdat-by-time.json", "w") as f:
    f.write(dumps(years_list))
```

```python
def plot(label, lam=None):
    lookup = t.labels.index(label)
    interesting = [vals[lookup-3] for year,vals in years_group]
    mi = max(interesting)
    print time[interesting.index(mi)],mi
    x = years_dict[time[interesting.index(mi)]].rows
    x.sort(key=lambda x: x[lookup], reverse=True)
    print x
    plt.plot(time, interesting)
    plt.title("%s with time" %(label))
    plt.xlim([1900,2011])
    plt.ylabel(label)
    plt.xlabel("time")
    if lam:
        lam(plt)
    plt.show()
```
